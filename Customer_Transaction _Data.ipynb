```python
#Merge two data
Data=pd.merge(df1,df2)
```

```python
Data.head(10)
```

```python
#RFM Calculation
#Groupby on CustomerID
rfm = Data.groupby('CustomerID').agg({'TransactionDate': ['max', 'count'],'TransactionAmount': 'sum'})
```

```python
rfm
```

```python
#calculate RFM
reference_date = Data['TransactionDate'].max() + pd.Timedelta(days=1)
reference_date
```

```python
rfm.columns = ['LastTransactionDate', 'Frequency', 'Monetary']
```

```python
rfm
```

```python
rfm['Recency'] = (reference_date - rfm['LastTransactionDate']).dt.days
```

```python
rfm
```

```python
#load RFM in DataFrame
df_rfm = rfm[['Recency', 'Frequency', 'Monetary']].reset_index()
```

```python
df_rfm.head(10)
```

```python
#Calculate RFM Score
df_rfm['R_score'] = pd.qcut(df_rfm['Recency'], 5, labels=[5,4,3,2,1]).astype(int)
df_rfm['F_score'] = pd.qcut(df_rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5]).astype(int)
df_rfm['M_score'] = pd.qcut(df_rfm['Monetary'], 5, labels=[1,2,3,4,5]).astype(int)
```

```python
df_rfm.head(10)
```

```python
#Combined RFM Score
df_rfm['Combined_RFM_Score'] = (df_rfm['R_score'].astype(str) + df_rfm['F_score'].astype(str) + df_rfm['M_score'].astype(str))
```

```python
df_rfm.head(10)
```

```python
#Counts of Combined RFM Score
RFM_score_count=df_rfm['Combined_RFM_Score'].value_counts()
RFM_score_count
```

```python
#Assign Segment Labels
def segment_row(r,f,m):
    if (r>=4) and (f>=4) and (m>=4):
        return 'Champions'
    elif (f>=4) and (r>=2):
        return 'Loyal Customer'
    elif (r>=4) and (2<=f<=3):
        return 'Potential Loyalist'
    elif (r<=2) and (f>=3):
        return 'At Risk'
    elif (m>=4) and (2<=f<=3) and (r>=3):
        return 'Big Spenders'
    elif (r==1) and (f<=2) and (m<=2):
        return 'Lost'
    else:
        return 'Others'

df_rfm['Assign_Segment'] = [segment_row(r,f,m) for r,f,m, in zip(df_rfm['R_score'],df_rfm['F_score'],df_rfm['M_score'])]
```

```python
df_rfm.head(10)
```

```python
#Showing Business Summary
print('\n===========================')
print('\n OVERALL BUSINESS SUMMARY')
print('\n===========================')

total_customers=df_rfm['CustomerID'].shape[0]
total_revenue=df_rfm['Monetary'].sum()
avg_revenue_per_customer=df_rfm['Monetary'].mean()

print('TOTAL CUSTOMERS:',  total_customers)
print('TOTAL REVENUE:',  total_revenue)
print('AVERAGE REVENUE PER CUSTOMER:', avg_revenue_per_customer) 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Customer Count in Each Segments
segment_counts = df_rfm['Assign_Segment'].value_counts().reset_index()
segment_counts.columns=['Assign_Segment','Customer_count']
segment_counts
```

```python
#Visualize(Bar Chart) Customer Count in Each Segments
plt.figure(figsize=(10,4))
sns.barplot(data=segment_counts,x='Assign_Segment',y='Customer_count',color='seagreen')
plt.title('CUSTOMER COUNT PER SEGMENT')
plt.xlabel('----SEGMENTS----')
plt.xticks(rotation=45)
plt.ylabel('---NUMBER OF CUSTOMERS---')
plt.show()
```

```python
#Revenue Contribution per Segment
segment_revenue = df_rfm.groupby('Assign_Segment')['Monetary'].sum().reset_index()
print('\n===========================')
print('REVENUE FROM EACH SEGMENT')
print('\n===========================')
segment_revenue
```

```python
#showing Total Revenue
total_revenue = segment_revenue['Monetary'].sum()
print('\n===========================')
print('TOTAL REVENUE')
print('\n===========================')
total_revenue
```

```python
#Percentage of Revenue Contribution per Segment
segment_revenue ['Revenue_Contribution_%'] = (segment_revenue['Monetary'] / total_revenue*100)
segment_revenue 
```

```python
#Round 2 of Percentage of Revenue Contribution
segment_revenue ['Revenue_Contribution_%'] = (segment_revenue['Monetary'] / total_revenue*100).round(2)
print('\n===========================')
print('REVENUE CONNTRIBUTION % BY SEGMENT')
print('\n===========================')
segment_revenue
```

```python
#Visualize(pie Chart) Percentage of Revenue Contribution per Segment
plt.figure(figsize=(4,4))
plt.pie(segment_revenue['Revenue_Contribution_%'],labels=segment_revenue['Assign_Segment'],autopct='%1.1f%%')
plt.title("-------REVENUE CONTRIBUTION PER SEGMENT-------")
plt.show()
```

```python
#Pareto Analysis
rfm_sorted=df_rfm.sort_values('Monetary', ascending=False)
rfm_sorted['CumRevenue']=rfm_sorted['Monetary'].cumsum()/rfm_sorted['Monetary'].sum()*100
x80=rfm_sorted.loc[rfm_sorted['CumRevenue']>=80,'CustomerID'].index[0] +1
pct_customers=x80/len(rfm_sorted)*100
print('\n=========================================================================')

print(f'\n TOP {pct_customers:.1f}% OF CUSTOMERS CONTRIBUTE ~80% OF TOTAL REVENUE')

print('\n=========================================================================')
```
